CLEANING DATA

```python
df.drop(['dmaj', 'insee', 'departement', 'commune', 'adresse_1', 'date_de_protection', 'precision_sur_la_protection', 'auteur', 'description', 'historique', 'affectataire', 'adresse', 'ancienne_region', 'reference', 'contact'], axis=1, inplace=True)
```

```python
df.drop(['longitude','latitude', 'coordonnees', 'coordonnees_ban'], axis=1, inplace=True)
```

```python
df['Longitude'] = df['coordonnees_finales'].str.split(',', expand=True)[0]
df['Latitude'] = df['coordonnees_finales'].str.split(',', expand=True)[1]
```

```python
df[['siecle']].value_counts()
```

```python
df['siecle_0'] = df['siecle'].apply(lambda x : re.findall('(\d*)..siècle', str(x)))
df['siecle_min'] = df['siecle_0'].apply(lambda x : min(x) if x else None)
```

```python
df["appellation_courante"] = df["appellation_courante"].apply(lambda x : np.where("Eglise" in x, "Eglise",x))
df["appellation_courante"] = df["appellation_courante"].apply(lambda x : np.where("Maison" in x, "Maison",x))
df["appellation_courante"] = df["appellation_courante"].apply(lambda x : np.where("Immeuble" in x, "Immeuble",x))
df["appellation_courante"] = df["appellation_courante"].apply(lambda x : np.where("Hôtel" in x, "Hôtel",x))
df["appellation_courante"] = df["appellation_courante"].apply(lambda x : np.where("Château" in x, "Château",x))
df["appellation_courante"] = df["appellation_courante"].apply(lambda x : np.where("Abbaye" in x, "Abbaye",x))
df["appellation_courante"] = df["appellation_courante"].apply(lambda x : np.where("Abri" in x, "Abri",x))
df["appellation_courante"] = df["appellation_courante"].apply(lambda x : np.where("Croix" in x, "Croix",x))
```
